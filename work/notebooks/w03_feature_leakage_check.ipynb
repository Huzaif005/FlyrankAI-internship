## 1. Build the feature vector

*Code that actually builds it — engineered features, categorical handling, fills.*

```python
import pandas as pd

# Load dataset
df = pd.read_csv(
    "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
)

# Select features available in the dataset
features = [
    "Pclass",
    "Sex",
    "Age",
    "Siblings/Spouses Aboard",
    "Parents/Children Aboard",
    "Fare"
]

X = df[features].copy()

# Fill missing values
X["Age"] = X["Age"].fillna(X["Age"].median())

# Convert categorical feature
X = pd.get_dummies(X, columns=["Sex"], drop_first=True)

# Target
y = df["Survived"]

print("Feature Matrix Shape:", X.shape)
print("\nFirst Five Rows:")
print(X.head())
```

## 2. Feature notes

| Feature | Meaning | Missing Values | Available Before Prediction? |
|----------|---------|---------------|------------------------------|
| Pclass | Passenger ticket class | No | Yes |
| Sex | Passenger gender | No | Yes |
| Age | Passenger age | Filled using median | Yes |
| Siblings/Spouses Aboard | Number of siblings/spouse aboard | No | Yes |
| Parents/Children Aboard | Number of parents/children aboard | No | Yes |
| Fare | Ticket fare | No | Yes |
| Embarked | Port of embarkation | Filled using mode | Yes |

All selected features are known before the survival outcome and are suitable for prediction.

```python
print("Missing Values Before Processing")
print(df[features].isnull().sum())

print("\nMissing Values After Processing")
print(X.isnull().sum())
```

## 3. The Leakage Hunt

I checked whether any selected feature directly reveals the target variable or contains information that would only be known after the survival outcome.

The target column **Survived** is excluded from the feature vector. No future information or label-derived columns are used, reducing the risk of data leakage.

```python
print("Feature Columns")
print(X.columns.tolist())

print("\nTarget Column")
print(y.name)

# Check whether target accidentally exists in features
print("\nLeakage Check")
print("Survived in Features:", "Survived" in X.columns)
```

## 4. What I Excluded and Why

| Excluded Field | Reason |
|----------------|--------|
| Survived | Target variable. Using it as a feature would cause data leakage. |
| Name | Mostly unique for each passenger and does not generalize well. |
| Ticket | Mostly unique ticket numbers with limited predictive value. |
| Cabin | Contains many missing values and incomplete information. |

These fields were excluded to reduce overfitting and prevent leakage while keeping only meaningful predictors.
