## 1. My Lane as an ML Task

This project is a **classification** task because the goal is to predict whether a passenger survived the Titanic disaster. The output has two possible classes: **Survived (1)** or **Did Not Survive (0)**. Classification is appropriate because the target variable is categorical.

```python
import pandas as pd

df = pd.read_csv("https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv")

print("Target Classes:")
print(df["Survived"].value_counts())
```

## 2. Target or Proxy

The target variable is **Survived**. This label comes from the observed historical outcome recorded in the dataset. A value of **1** means the passenger survived, while **0** means the passenger did not survive. Therefore, the model will learn from actual historical observations rather than a manually defined rule.

```python
print("Target Column:", "Survived")

print(df["Survived"].value_counts())

print("\nPercentage Distribution")
print(df["Survived"].value_counts(normalize=True) * 100)
```

## 3. Success Metric

The primary evaluation metric for this classification task is **Accuracy** because it measures the percentage of correctly classified passengers. A model accuracy above **80%** would indicate good performance on this dataset. Additional metrics such as Precision, Recall, and F1-score may also be considered if class imbalance becomes important.

```python
from sklearn.metrics import accuracy_score

print("Selected Evaluation Metric: Accuracy")
print("Target Accuracy: Above 80%")
```

## 4. Unit of Analysis

One row in the dataframe represents **one Titanic passenger**. Each row contains information such as passenger class, age, gender, fare, and survival status. These features are used to learn patterns associated with survival.

```python
print("Shape of Dataset:", df.shape)

print("\nFirst Five Rows")
display(df.head())
```

## 5. Why ML Beats a Fixed Rule

Passenger survival depends on several factors such as age, gender, passenger class, fare, and their interactions. These relationships are too complex to represent with a simple if-else rule. Machine learning can discover patterns from historical data and combine multiple features to make better decision-support predictions than a fixed rule.

```python
features = ["Pclass", "Sex", "Age", "Fare"]

print("Features Used:")
print(features)

print("\nFeature Summary")
print(df[features].describe(include="all"))
```
