## 1. Distributions

*Look before deciding: distributions of your key fields. Note the heavy tails.*

```python
import pandas as pd

# Load dataset
df = pd.read_csv("https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv")

# Summary statistics
print(df.describe())

# Distribution of categorical variables
print("\nPassenger Class Distribution:")
print(df["Pclass"].value_counts())

print("\nGender Distribution:")
print(df["Sex"].value_counts())

print("\nSurvival Distribution:")
print(df["Survived"].value_counts())
```

## 2. Signal Tests

### Signal 1: Female passengers had a higher survival rate.
**Verdict:** CONFIRMED

### Signal 2: First-class passengers had a higher survival rate.
**Verdict:** CONFIRMED

### Signal 3: Older passengers had lower survival rates.
**Verdict:** MIXED

The first two signals are clearly supported by the observed data. The relationship between age and survival is weaker and varies across passenger groups.

```python
# Signal 1
print("Survival Rate by Gender")
print(df.groupby("Sex")["Survived"].mean())

# Signal 2
print("\nSurvival Rate by Passenger Class")
print(df.groupby("Pclass")["Survived"].mean())

# Signal 3
print("\nAverage Age by Survival")
print(df.groupby("Survived")["Age"].mean())
```

## 3. Flag-Linked Test

One important assumption is that passenger class is associated with survival. The analysis compares survival rates across passenger classes to determine whether this assumption is supported by the observed data.

The results indicate that first-class passengers had a noticeably higher survival rate than second- and third-class passengers. Therefore, this signal is supported by the available data.

```python
survival_by_class = df.groupby("Pclass")["Survived"].mean()

print(survival_by_class)
```

## 4. What This Means in Practice

The analysis shows that passenger class and gender are stronger indicators of survival than age alone. These observed relationships can be useful as input features for a machine learning model, but they should be treated as decision-support rather than proof of causation. Additional factors not included in the dataset may also have influenced survival.
